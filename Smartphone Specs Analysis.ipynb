```python
### Spliting the Connectivity column
def extract_connectivity_parts(row):
    items = str(row).split(', ')
    
    sim_type = None
    networks = []
    enhancers = []
    
    for item in items:
        if 'Sim' in item:
            sim_type = item
        elif item in ['3G', '4G', '5G']:
            networks.append(item)
        else:
            enhancers.append(item)

    return pd.Series([
        sim_type, 
        ', '.join(networks), 
        ', '.join(enhancers)
    ])

df[['SIM_Type', 'Networks', 'Enhancers']] = df['Connectivity'].apply(extract_connectivity_parts)
```

```python
#The function will split a column after every comma into the specified columns
def split_columns(table, column, char,*args):
    table[[*args]] = table[column].astype(str).str.split(char, n=len(args)-1, expand=True)
    table.drop(columns=[column], inplace=True)
    

split_columns(df,'Processor',',','Chipset', 'Core_Type', 'Clock_Speed')
```

```python
def drop_terms(table,column,*args):
    words_to_remove = [*args]
    
    for word in words_to_remove:
        table[column] = table[column].str.replace(word, '', regex=False)
```
